```python
categorical = list(df.select_dtypes(include='object'))
print(f"categories: {categorical}")
```

```python
from sklearn.model_selection import train_test_split

y = df.tsales
X = df.drop(columns=['hourspw', 'hoursw', 'nfull', 'naux', 'npart', 'tsales'])
X = sm.add_constant(X)

print(X.head(2), y.head(2))

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=1502
)

X_train.head(2)
```

```python
model = sm.OLS(y_train, X_train).fit()
print(model.summary())
```

```python
preds = model.predict(X_test)
preds[:4]
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
print(f"MAE: {round(mean_absolute_error(y_test, preds), 1)}")
print(f"MSE: {round(np.sqrt(mean_squared_error(y_test, preds)), 1)}")
```
