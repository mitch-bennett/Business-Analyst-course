```python
# libraries
import pandas as pd
import seaborn as sns
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
```

```python
# load data
df = pd.read_csv("titanic.csv")
df.describe()
```

```python
df.describe()
```

```python
df.hist(figsize=(15,10))
```

```python
# remove outliers
print(len(df))
df = df[(df['Parents/Children Aboard'] <= 3) & (df['Fare'] <= 300) & (df['Siblings/Spouses Aboard'] <= 5)]
print(len(df))
```

```python
df.hist(figsize=(15,10))
```

```python
# one hot code sex to int
# df['Sex'] = np.where(df['Sex'] == 'male', 1, 0)
df = df.drop(columns=['Name'])

#Transform character variables
df = pd.get_dummies(data = df, drop_first=True)

df.head()
```

```python
# correlation matrix
sns.heatmap(data=df.corr(),
            annot=True,
            fmt='.1g',
            center=0,
            cmap='coolwarm',
            linecolor='black',
            linewidths=1)
```

```python
# set up variables
y = df.Survived
X = df.iloc[:,2:]
X = sm.add_constant(X)
X.head(3)
```

```python
# train/test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size=0.2,
                                                    random_state=1502)

# convert binary values to float
X_train = X_train.astype(float)
X_test = X_test.astype(float)
y_train = y_train.astype(float)
y_test = y_test.astype(float)
```

```python
model = sm.Logit(y_train, X_train).fit()
print(model.summary())
```

```python
preds = model.predict(X_test)
preds = np.where(preds > 0.5, 1, 0)
print(preds[:4])
```

```python
y_test.mean()
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test, preds))
tn, fp, fn, tp = confusion_matrix(y_test, preds).ravel()
print(f"True negatives: {tn}\nFalse positives: {fp}\nFalse negatives: {fn}\nTrue positives: {tp}")
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, preds))
```
